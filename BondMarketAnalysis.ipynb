```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import scipy as sp
import matplotlib.pyplot as  plt
import yfinance as yf
from pandas import Series,DataFrame
from numpy.random import randn
from scipy import stats
from datetime import datetime
sns.set_style('whitegrid')
%matplotlib inline
```

```python
#reads data of 13 month,5 year,10 year and 30 year US treasury bonds over 1 year in dataframes named after their symbols
end=datetime.now()
start=datetime(end.year-1,end.month,end.day)
bondlist=['^IRX','^FVX','^TNX','^TYX']
for bond in bondlist:
    Bond=bond[1:]
    globals()[Bond]=yf.download(bond,start,end)
```

```python
#reads closing price data of 13 month,5 year,10 year and 30 year US treasury bonds over 1 year
closing_df=yf.download(bondlist,start,end)['Adj Close']
```

```python
#plots yield of 13 month,5 year,10 year and 30 year US treasury bonds over 1 year
closing_df.plot(legend=True,figsize=(10,4))
```
